```python
### Lists

# 1. Reverse a list in-place
def reverse_list(lst):
    left, right = 0, len(lst) - 1
    while left < right:
        lst[left], lst[right] = lst[right], lst[left]
        left += 1
        right -= 1
    return lst

# 2. Find the second largest element
def second_largest(lst):
    unique_lst = list(set(lst))
    unique_lst.sort()
    return unique_lst[-2] if len(unique_lst) > 1 else None

# 3. Remove all occurrences of an element
def remove_occurrences(lst, val):
    return [x for x in lst if x != val]

# 4. Check if two lists have the same elements (regardless of order)
def same_elements(lst1, lst2):
    return set(lst1) == set(lst2)

# 5. Flatten a nested list
def flatten_list(nested_lst):
    flat_list = []
    for sublist in nested_lst:
        if isinstance(sublist, list):
            flat_list.extend(flatten_list(sublist))
        else:
            flat_list.append(sublist)
    return flat_list

# 6. Rotate a list by n positions
def rotate_list(lst, n):
    n = n % len(lst)
    return lst[n:] + lst[:n]

# 7. Find common elements between two lists
def common_elements(lst1, lst2):
    return list(set(lst1) & set(lst2))

# 8. Insert an element at a specific index
def insert_at(lst, index, val):
    lst.insert(index, val)
    return lst

# 9. Delete an element at a specific index
def delete_at(lst, index):
    if 0 <= index < len(lst):
        lst.pop(index)
    return lst

lst=[1,2,3,4,5]
delete_at(lst,2)

# 10. Create a list of the first n Fibonacci numbers
def fibonacci(n):
    fib = [0, 1]
    for _ in range(n - 2):
        fib.append(fib[-1] + fib[-2])
    return fib[:n]

### Tuples

# 1. Access the last element
def last_element(tpl):
    return tpl[-1]

# 2. Find the index of a specific element
def index_of(tpl, val):
    return tpl.index(val) if val in tpl else -1

# 3. Check if a tuple contains a specific element
def contains(tpl, val):
    return val in tpl

# 4. Concatenate two tuples
def concatenate_tuples(tpl1, tpl2):
    return tpl1 + tpl2

# 5. Repeat a tuple n times
def repeat_tuple(tpl, n):
    return tpl * n

# 6. Convert a list to a tuple
def list_to_tuple(lst):
    return tuple(lst)

# 7. Convert a tuple to a list
def tuple_to_list(tpl):
    return list(tpl)

# 8. Find the length of a tuple
def tuple_length(tpl):
    return len(tpl)

# 9. Unpack a tuple into multiple variables
def unpack_tuple(tpl):
    return tpl  # Simply return it for demonstration

# 10. Create a tuple of even numbers from a given list
def even_numbers_tuple(lst):
    return tuple(x for x in lst if x % 2 == 0)

### Sets

# 1. Find intersection of two sets
def intersection(set1, set2):
    return set1 & set2

# 2. Find union of two sets
def union(set1, set2):
    return set1 | set2

# 3. Find difference between two sets
def difference(set1, set2):
    return set1 - set2

# 4. Check if a set is a subset of another
def is_subset(set1, set2):
    return set1.issubset(set2)

# 5. Add an element to a set
def add_element(s, val):
    s.add(val)
    return s

# 6. Remove an element from a set (handle the case where it's not present)
def remove_element(s, val):
    s.discard(val)
    return s

# 7. Convert a list to a set (remove duplicates)
def list_to_set(lst):
    return set(lst)

# 8. Find symmetric difference
def symmetric_difference(set1, set2):
    return set1 ^ set2

# 9. Create a set from a string
def string_to_set(s):
    return set(s)

# 10. Find the number of common elements between two sets
def common_elements_count(set1, set2):
    return len(set1 & set2)

### Dictionaries

# 1. Add a new key-value pair
def add_pair(dct, key, value):
    dct[key] = value
    return dct

# 2. Remove a key-value pair
def remove_key(dct, key):
    dct.pop(key, None)
    return dct

# 3. Check if a key exists
def key_exists(dct, key):
    return key in dct

# 4. Get value associated with a key
def get_value(dct, key):
    return dct.get(key, None)

# 5. Iterate through keys
def iterate_keys(dct):
    return list(dct.keys())

# 6. Iterate through values
def iterate_values(dct):
    return list(dct.values())

# 7. Iterate through key-value pairs
def iterate_items(dct):
    return list(dct.items())

# 8. Merge two dictionaries
def merge_dicts(dct1, dct2):
    dct1.update(dct2)
    return dct1

# 9. Create a dictionary with word frequencies
def word_frequencies(text):
    words = text.split()
    return {word: words.count(word) for word in set(words)}

# 10. Find the most frequent value in a dictionary
def most_frequent_value(dct):
    from collections import Counter
    return Counter(dct.values()).most_common(1)[0][0] if dct else None

```
