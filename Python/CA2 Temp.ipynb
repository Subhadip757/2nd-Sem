```python
# Scenario Analyzing Sales Data of an Online BookStore

# You are working as a data analyst for an online bookstore. 
# You have been provided with a CSV file named book_sales_csv
# containing the following information for each transaction:

#     *order_id: Unique identifier for each other.
#     *book_title: Title of the book purchased
#     *author: Author of the book
#     *genre: Genre of the book (eg., Fiction, Mystery, Science Fiction, Non-Fiction).
#     *quantity: Number of copies of the book purchased in the order.
#     *price: Price of a single copy of the book (in INR)
#     *order_date: Date when the order was placed (in YYYY-MM-DD format).


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load and Insert Data (Pandas)
try:
    sales_df = pd.read_csv('book_sales.csv')
    print("First 5 rows of the DataFrame: ")
    print(sales_df.head())
    print("\nData Types of each Column:")
    print(sales_df.dtypes)
except FileNotFoundError:
    print("Error: The file 'book_sales.csv' was not found. Please make sure the destination")
    exit()
```

```python
sales_df['total_revenue'] = sales_df['quantity'] * sales_df['price']
total_revenue = sales_df['total_revenue'].sum()
print(f"\nTotal revenue generated by the bookstore: ₹{total_revenue:.2f}")
```

```python
# 3. Analyze Genre Distribution (Pandas & Matplotlib):
genre_counts = sales_df['genre'].value_counts()
print("\nNumber of orders for each genre: ")
print(genre_counts)

plt.figure(figsize=(10,6))
genre_counts.plot(kind='bar', color='skyblue')
plt.xlabel("Genre")
plt.ylabel("Number of Orders")
plt.title("Distribution of Orders Across Book Genres")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
sales_df['order_date'] = pd.to_datetime(sales_df['order_date'], format='%m-%m-%y')
daily_revenue = sales_df.groupby('order_date')['total_revenue'].sum()

plt.figure(figsize=(12,6))
```
